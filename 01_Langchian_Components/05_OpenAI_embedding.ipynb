```python
from langchain.embeddings import OpenAIEmbeddings
embeddings = OpenAIEmbeddings(model="text-embedding-3-small") # dim = 1536
# we can change the dim also 
# embeddings = OpenAIEmbeddings(model="text-embedding-3-small" , dim = 1024) 
embeddings
```

```python
text = "This is a test sentence."
vector = embeddings.embed_query(text)
print(f"Vector for '{text}': {vector}")
vector
```

```python
len(vector)  # Length of the vector
```

```python
## Text loader

from langchain_community.document_loaders import TextLoader
loader = TextLoader("speech.txt" , encoding="utf-8")
test_doc = loader.load()
test_doc
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
final_docs = text_splitter.split_documents(test_doc)
final_docs
```

```python
## Vector Embedding And Vector StoreDB
from langchain.vectorstores import Chroma
db = Chroma.from_documents(final_docs, embeddings, persist_directory="db")
db
```

```python
# Retruve the results from query vector store db
query = "What is the main topic of the speech?"
docs = db.similarity_search(query)
print(docs)
```
