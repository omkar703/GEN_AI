# Ollama 

ollama supports embedding models , making it possible to build retrieval systems that are more relevant to the user's query.

```python
from langchain_community.embeddings import OllamaEmbeddings
```

```python
embeddings = (
    OllamaEmbeddings(
        # model="llama3.2:latest", # by default llama2
        model = "mxbai-embed-large:latest",
        )
)
```

```python
embeddings
```

```python
r1= embeddings.embed_query([
    "Alpha is the first letter of the Greek alphabet.",
    "Beta is the second letter of the Greek alphabet.",
])
```
