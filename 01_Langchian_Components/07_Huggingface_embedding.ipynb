# Sentence Transformer Embeddings
    hugging face sentence transformer embeddings is a python freamwork for generating embeddings from text using pre-trained models available on Hugging Face. text and image embeddings are supported. it is a part of the Hugging Face ecosystem and can be used with various models for tasks like semantic search, clustering, and classification.

```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
os.environ['HF_TOKEN'] = os.getenv("HF_TOKEN")
```

```python
from langchain_huggingface import HuggingFaceEmbeddings
embeddings = HuggingFaceEmbeddings(model_name = "all-MiniLM-L6-v2")
```

```python
text = "This is atest document"
query_result = embeddings.embed_query(text)
query_result
```

```python
len(query_result)
```

```python
doc_result = embeddings.embed_documents([text , "This is not a test document"])
doc_result
```
